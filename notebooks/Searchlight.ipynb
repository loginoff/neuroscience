We load in the data for a single subject using our own little utility function.
By default it load all the runs for the subject and also find a corresponding class label
for each sample ("targets" attribute)

Initial preprocessing of the data consists of linear detrending and zscoring.
Both are done in place.

```python
ds=load_openfmri_ds('../data/ds107','sub001')
poly_detrend(ds, polyord=1, chunks_attr='chunks')
zscore(ds, param_est=('targets',['rest']))
ds=ds[ds.sa.targets!='rest']
create_folds(ds,6)
```

```python
ds.sa['half'] = map(lambda x: 1 if x < 5 else 2, ds.sa.folds)
```

```python
def group_count(data):
    def counter(acc, curr):
        if curr in acc:
            acc[curr]+=1
        else:
            acc[curr]=1
        return acc
    
    return reduce(counter, data,{})

def create_folds(ds, nrfolds, targets_attribute='targets', folds_attribute='folds'):
    counts={}
    def calc_fold(elem):
        if elem in counts:
            counts[elem]+=1
            return counts[elem] % nrfolds
        else:
            counts[elem]=1
            return 1%nrfolds
    ds.sa[folds_attribute]=map(calc_fold, ds.sa[targets_attribute])
```

```python
for radius in [2, 3, 4, 5, 6, 10]:
    clf = LinearCSVMC()
    #clf = kNN()
    partitioner = HalfPartitioner(attr='half')
    cvte = CrossValidation(clf,partitioner,errorfx=lambda p, t: np.mean(p == t))
    sl = sphere_searchlight(cvte, radius=radius, nproc=40)
    res=sl(ds)
    np.save('linearCSVMC_radius_%d' % radius, res.samples)
```

```python
def get_spheres(ds, centers, radius=10):
    def euclid_distance(x,y):
        return math.sqrt((x[0]-y[0])**2 + (x[1]-y[1])**2 + (x[2]-y[2])**2)
    def compare_to_center(voxel):
        for c in centers:
            if euclid_distance(voxel, c) <= radius:
                return True
        return False
    return filter(compare_to_center, ds.fa.voxel_indices)
    
```

```python
sphrs=get_spheres(ds,centers,5)
len(sphrs)
```
