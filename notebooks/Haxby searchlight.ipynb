```python
import numpy as np
%pylab --no-import-all inline

from mvpa2.mappers.detrend import poly_detrend
from mvpa2.mappers.zscore import ZScoreMapper
from mvpa2.mappers.zscore import zscore
from util.openfmri import load_openfmri_ds

from mvpa2.clfs.knn import kNN
from mvpa2.clfs.distance import one_minus_correlation
from mvpa2.clfs.svm import LinearCSVMC
from mvpa2.generators.partition import NFoldPartitioner
from mvpa2.generators.partition import HalfPartitioner
from mvpa2.measures.base import CrossValidation
from mvpa2.measures.searchlight import sphere_searchlight
import math
```

```python
ds=load_openfmri_ds('../data/haxby','sub001', TR=2.5, mask='../data/haxby/sub001/masks/orig/vt.nii.gz')
poly_detrend(ds, polyord=1, chunks_attr='chunks')
zscore(ds, param_est=('targets',['rest']))
ds=ds[ds.sa.targets!='rest']
```

```python
ds.sa['half'] = map(lambda x: 1 if x < 11 else 2, ds.sa.chunks)
```

```python
from mvpa2.datasets.mri import map2nifti
img=map2nifti(ds,ds.samples[0])
img.to_filename('haxby_anatomy.nii.gz')
img.shape
```

```python
clf = LinearCSVMC()
partitioner = HalfPartitioner(attr='half')
cvte = CrossValidation(clf,partitioner,errorfx=lambda p, t: np.mean(p == t))
sl = sphere_searchlight(cvte, radius=4, nproc=40)
res=sl(ds)
np.save('Haxby_searchlight_radius_4', res.samples)
```

```python
# little helper function to plot dissimilarity matrices
def plot_mtx(mtx, labels, title):
    plt.figure()
    plt.imshow(mtx, interpolation='nearest')
    plt.xticks(range(len(mtx)), labels, rotation=-45)
    plt.yticks(range(len(mtx)), labels)
    plt.title(title)
    plt.clim((0,1))
    plt.colorbar()
```

```python
# basic ROI RSA -- dissimilarity matrix for the entire ROI
from mvpa2.measures import rsa
dsm = rsa.PDist(square=True)
res = dsm(ds)
plot_mtx(res, ds.sa.targets, 'ROI pattern correlation distances')
```
